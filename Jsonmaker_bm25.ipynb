```python
import pandas as pd
import numpy as np
import requests
```

simple request; searching for 'toyota':


```python
queryfile = "./queries-v2.txt"
qlist = []

with open(queryfile, 'r') as f:
    for line in f:
        sep = line.split("\t")
        query = sep[1]
        query = query.rstrip("\n")
        #adding plus sign instead of space to append to queries
        query = query.replace(" ", "+")        
        qlist.append(query)
qlist
```

doing it for all of the queries from the list:

```python
import json
import os
for i in qlist:
    if not os.path.exists('query_results_bm25/'+i):
        os.makedirs('query_results_bm25/'+i)
```
